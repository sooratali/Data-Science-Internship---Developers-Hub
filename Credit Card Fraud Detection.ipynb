```python
import numpy as np

def test_transaction_input(model, scaler):
    print("\n--- Fraud Detection Test Interface ---")
    print("Enter the 30 transaction features (as required by the model).")

    inputs = []
    for i in range(30):
        val = float(input(f"Feature {i + 1}: "))
        inputs.append(val)

    input_array = np.array(inputs).reshape(1, -1)
    input_scaled = scaler.transform(input_array)

    prediction = model.predict(input_scaled)
    print("\n🔍 Prediction:", "🚨 Fraudulent Transaction" if prediction[0] == 1 else "✅ Legitimate Transaction")
```
